```python
# Program that counts the frequency of each character in a string
'''
approach:
utilizing dicts to store frequency of each character, then iterate over each character in the input string
aur fir print kardo
'''
a = input("Enter a string: ")
frequency = {}
for char in a:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1
for char, count in frequency.items():
    print(f"'{char}': {count}")
```
